```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Charger les données depuis le fichier Excel
file_path = 'C:/Documents/EPFL/MA4/Projet_de_semestre/br_100_s1_1a_P-V 1V_1#1.xls'
data = pd.read_excel(file_path) 

# Tracer un graphique simple (remplacer 'Colonne1' et 'Colonne2' par les noms réels des colonnes)
plt.figure(figsize=(10, 5))
plt.plot(data['Time'], data['Vforce'], marker='o')
plt.title('Titre du Graphique')
plt.xlabel('Time')
plt.ylabel('Force')
plt.grid(True)
plt.show()
```
